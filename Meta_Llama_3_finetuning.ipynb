```python
import json
import torch
from transformers import (AutoTokenizer,
                          AutoModelForCausalLM,
                          BitsAndBytesConfig,
                          pipeline)
```

### **HF Account Config**

```python
config_data = json.load(open("config.json"))
HF_TOKEN = config_data["HF_TOKEN"]
```

```python
print(HF_TOKEN)
```

```python
model_name = "meta-llama/Meta-Llama-3-8B"
```

### **Quantisation Configuration**

```python
bnb_config = BitsAndBytesConfig(
    load_in_4bit = True,
    bnb_4bit_use_double_quant = True,
    bnb_4bit_quant_type = "nf4",
    bnb_4bit_compute_dtype = torch.bfloat16
)
```

### **Loading the tokenizer and LLM**

```python
tokenizer = AutoTokenizer.from_pretrained(model_name, token=HF_TOKEN)
tokenizer.pad_token = tokenizer.eos_token
```

```python
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    device_map = "auto",
    quantization_config = bnb_config,
    token=HF_TOKEN
)
```

```python
text_generator = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=128
)
```

```python
def get_response(prompt):
  sequences = text_generator(prompt)
  gen_text = sequences[0]["generated_text"]
  return gen_text
```

```python
prompt = "What is agent based learning?"
```

```python
llama3_response = get_response(prompt)
```

## **Extract data from PDF**

### **Extract text from PDF**

```python
!pip install pymupdf
```

```python
import fitz

with fitz.open('TextPdf.pdf') as pdf:
    text = ''
    for page in pdf:
        text += page.get_text()
print(text)
```

### **Text Preprocessing**

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

text = text.lower()
text = re.sub(r'[^a-z\s]', '', text)
text = re.sub(r'\s+', ' ', text).strip()

nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
text = ' '.join([word for word in text.split() if word not in stop_words])

#nltk.download('punkt')
#tokens = word_tokenize(text)
#print(tokens)
print(text)
```

### **Identify medical entities like diseases, deficiencies**

```python
import spacy
from scispacy.linking import EntityLinker
import scispacy

nlp = spacy.load("en_core_sci_md")
doc = nlp(text)

for ent in doc.ents:
    print(ent.text)
```

#### Custom Keyword Search

```python
diseases = ["diabetes", "hypertension", "anemia", "vitamin D deficiency", "alzheimers"]
mentioned_diseases = []

for disease in diseases:
    if disease.lower() in text.lower():
        mentioned_diseases.append(disease)

print("Mentioned Diseases/Deficiencies:", mentioned_diseases)
```

### **Extract text from images embedded in PDF**

```python
!pip install pymupdf
!pip install pytesseract
!pip install pillow
```

```python
import fitz
import io
from PIL import Image

pdf_path = 'pd_1.pdf'
pdf_document = fitz.open(pdf_path)

for page_number in range(len(pdf_document)):
    page = pdf_document.load_page(page_number)
    image_list = page.get_images(full=True)

    if image_list:
        for img_index, img in enumerate(image_list):
            xref = img[0]
            base_image = pdf_document.extract_image(xref)
            image_bytes = base_image["image"]
            image = Image.open(io.BytesIO(image_bytes))
            image_path = f"extracted_image_page{page_number+1}_img{img_index+1}.png"
            image.save(image_path)
            image.show()
```
